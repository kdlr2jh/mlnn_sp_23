```python
import pandas as pd

beans = pd.read_excel('Dry_Bean_Dataset/Dry_Bean_Dataset.xlsx')

beans.head()
```

```python
from sklearn import preprocessing

enc = preprocessing.OrdinalEncoder()
```

```python
beans_copy = beans.copy()

enc.fit(beans_copy[["Class"]])
enc.categories_
```

```python
beans_copy["Class"] = enc.fit_transform(beans_copy[["Class"]])
beans_copy.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(beans_copy.drop(['Class'], axis = 1), beans_copy.Class, test_size=.25)
#beans.shape, train.shape, test.shape
x_train.head()
```

```python
y_train.head()
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=7, random_state=1, n_init='auto')
km.fit(x_train)
```

```python
set(km.labels_)
```

```python
x_train['cluster'] = km.predict(x_train)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['font.size'] = 14
```

```python
fig = plt.figure()
plt.scatter(x_train.MajorAxisLength, x_train.MinorAxisLength, c=x_train.cluster, s=5, cmap='Paired')

plt.xlabel('Major Axis Length')
plt.ylabel('Minor Axis Length')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(x_train, km.labels_, metric='euclidean')
```

```python
K = range(2, 8)
fits = []
score = []

for k in K:
    model = KMeans(n_clusters = k, random_state = 1, n_init='auto')
    value = model.fit(x_train)
    
    fits.append(value)
    
    score.append(silhouette_score(x_train, model.labels_, metric='euclidean'))
```
